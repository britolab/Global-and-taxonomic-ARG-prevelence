```python
acc_accession=sequencing_data[~sequencing_data["NCBI_accession"].isna()]
first_acc=acc_accession.copy().drop_duplicates("study_name")
single_acc=[]

for acc in first_acc["NCBI_accession"]:
    single_acc.append(acc.split(";")[0])
first_acc["single_acc"]=single_acc
```

```python
#making a new list so that a single line for each accession number

# no need to download the brito genomes
no_brito=acc_accession[acc_accession["study_name"]!="BritoIL_2016"]
updated_list=[]
for index, row in no_brito.iterrows():
    accessions=row["NCBI_accession"].split(";")
    for acc in accessions:
        new_entry=[row["study_name"], row["sample_id"], acc]
        updated_list.append(new_entry)
melted_df=pd.DataFrame(updated_list, columns=["study_name","sample_id","single_accession"])
```

```python
# actually I do want the Brito samples
only_brito=acc_accession[acc_accession["study_name"]=="BritoIL_2016"]
updated_list=[]
for index, row in only_brito.iterrows():
    accessions=row["NCBI_accession"].split(";")
    for acc in accessions:
        new_entry=[row["study_name"], row["sample_id"], acc]
        updated_list.append(new_entry)
melted_df=pd.DataFrame(updated_list, columns=["study_name","sample_id","single_accession"])
melted_df.to_csv("brito_accessions_list.txt", sep='\t', index=False, header=False)
```

## Used the list of accessions in "wget_files.sh" to get the FTP addresses
* 1771 of the accessions did not have a url

```python
import math

ftp_df=pd.read_csv("all_urls.txt",sep='\t')
counts=[]
for item in ftp_df["ftp_urls"]:
    if type(item)==str:
        num_urls=len(item.split(";"))
        counts.append(num_urls)
    else: counts.append(0)
ftp_df["url_counts"]=counts
ftp_df.to_csv("ftp_urls_with_counts.txt",sep='\t')
```

```python
brito_ftp_df=pd.read_csv("brito_urls.txt",sep='\t')
counts=[]
for item in brito_ftp_df["ftp_urls"]:
    if type(item)==str:
        num_urls=len(item.split(";"))
        counts.append(num_urls)
    else: counts.append(0)
brito_ftp_df["url_counts"]=counts
brito_ftp_df.to_csv("brito_ftp_urls_with_counts.txt",sep='\t')
```

```python
ftp_df.groupby(["study_name","url_counts"]).size().reset_index().to_csv("summary_of_url_downloads.txt",sep='\t')
```

### After collecting all the FTP links I'm left with 14,367 samples
* I have to remove 177 samples from NielsenHB_2014 that are also in LeChatelierE_2013
* remove the milk becasue only 8 samples
* filter samples with read lengths lower than 70 (min reccommended lengeth for bwa mem)

```python
samples=sequencing_ftp_data.drop_duplicates(["study_name","sample_id"])
print(str(len(samples))+" samples with ftp link")

# remove the milk sampls
sample_filter=samples[samples["body_site"]!="milk"]
print(str(len(sample_filter))+" after removing milk samples")

# remove samples with read lengths less than 70
sample_filter=sample_filter[sample_filter["median_read_length"]>=70]
print(str(len(sample_filter))+" samples with median read lengths over 70bp")
#for in in range(1000000, )
```

### Finally export two lists for the main batch and then the ones with massive depth to save for later
* samples with more than 100 million reads will be set aside

```python
selected_samples=pd.merge(left=read_filtered[["study_name","sample_id"]], right=sequencing_ftp_data, on=["study_name","sample_id"], how="left")
selected_samples["study_sample"]=selected_samples["study_name"]+"-"+selected_samples["sample_id"]
high_samples=selected_samples[selected_samples["number_reads"]>100000000]
low_samples=selected_samples[selected_samples["number_reads"]<=100000000]
print(len(high_samples))
print(len(low_samples))
selected_samples.drop_duplicates(["study_sample"])[["study_sample"]].to_csv("accession_lists/00_all_read_samples.txt",sep='\t', index=False, header=False)
#low_samples.drop_duplicates(["study_sample"])[["study_sample"]].to_csv("accession_lists/01_low_read_samples.txt",sep='\t', index=False, header=False)
high_samples.drop_duplicates(["study_sample"])[["study_sample"]].to_csv("accession_lists/02_high_read_samples.txt",sep='\t', index=False, header=False)
```

```python
ftp_for_export=selected_samples.groupby(['study_name','sample_id','study_sample'])['ftp_urls'].apply(lambda x: ';'.join(x)).reset_index()
for index, row in ftp_for_export.iterrows():
    export_list=[]
    for url in row["ftp_urls"].split(";"):
        file_name=url.split("/")[-1].split(".fastq.gz")[0] 
        export_list.append([row["study_sample"], row["study_name"], row["sample_id"], file_name, url])
    pd.DataFrame(export_list).to_csv("accession_lists/"+row["study_sample"]+"_ftp_list.txt",sep='\t', index=False, header=False)
```
